```python
with open('bible.txt', 'r') as file:
    text = file.read()
```

```python
print(text[:1000])
```

```python
import re

text = text.lower()
symbols = set(text)
exclude = '[:0123456789*;,()-\']'

data = re.sub('?'.join(exclude), '', text)
data = re.sub('\t?\r?\n', ' ', data)

words = data.replace('.', '').split(' ')
frequencies = {word: words.count(word)for word in set(words)}
n_words = len(words)

print(data[:1000])
```

```python
top_100k = sorted(frequencies.items(), key=lambda x: -x[1])[:100_000]
top_100k
```

```python
import tensorflow_text as tft


splitter = tft.RegexSplitter('\.')
sentences = splitter.split(data)
splitter = tft.RegexSplitter(' ')
words = splitter.split(sentences)[0]
words
```

```python
import tensorflow as tf


tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=10000)
tokenizer.fit_on_texts([text])
```

```python
window = lambda seq, size: [seq[i:i+size] for i in range(seq.shape[0] - size)]

def flatten(l):
    x = []
    for i in l:
        x.extend(i)
    return x

size = 3
bags = [window(seq, size + 1 + size) for seq in words]
bags = flatten(bags)
pairs = [(word, bag[size]) for bag in bags for word in tf.concat([bag[:size], bag[size+1:]], axis=0)]
```

```python
pairs
```

```python
dataset = tf.data.Dataset.from_tensor_slices(pairs)
dataset
```
