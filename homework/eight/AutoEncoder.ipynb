```python
def preprocess(data):
    # DROPING
    data = data.map(lambda img, target: (img))
    # DATA TYPE
    data = data.map(lambda img: (tf.cast(img, tf.float32)))
    # NORMALIZE
    data = data.map(lambda img: (img/255.))
    # DUPING AND NOISE
    level = 0.5
    data = data.map(lambda img: ((tf.random.normal(img.shape) + \
                    tf.random.uniform(img.shape, -1, 1)**2) * level + img, img))
    # DATAFLOW PREP
    data = data.cache()
    data = data.shuffle(1000, seed=42)
    data = data.batch(32)
    data = data.prefetch(10)
    return data


train = train_raw.apply(preprocess)
test  = test_raw.apply(preprocess)
```

```python
import datetime
from tqdm import tqdm



class Model(tf.keras.Model):
    NAME = 'NAMELESS_MODEL'
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.is_setup = False
        self.loss_metric = tf.keras.metrics.Mean(name="loss")
        self.prev_epochs = 0
    
    
    def reset_metrics(self):
        for metric in self.metrics:
            metric.reset_state()
    
    
    def setup(self):
        if self.is_setup:
            return
        # DEFINE PATHSf
        current_time = datetime.datetime.now().strftime("%Y.%m.%d-%H:%M:%S")
        train_log_path = f"logs/{self.NAME}{current_time}/train"
        val_log_path = f"logs/{self.NAME}{current_time}/val"
        # CONSTRUCT WRITERS
        self.writer_train = tf.summary.create_file_writer(train_log_path)
        self.writer_val   = tf.summary.create_file_writer(val_log_path)
        self.is_setup = True
    
    
    @tf.function
    def call(self, x, training=False):
        for layer in self.layers:
            x = layer(x, training=training)
        return x
    
    
    @tf.function
    def step(self, X, T, training=True):
        if training:
            with tf.GradientTape() as tape:
                L = self.loss(X, T)
            gradient = tape.gradient(L, self.trainable_weights)
            self.optimizer.apply_gradients(zip(gradient, self.trainable_weights))
        else:
            L = self.loss(X, T)
        self.loss_metric.update_state(L)
        return {metric.name: float(metric.result()) for metric in self.metrics}


    def train(self, train, test, epochs):
        training_metrics = {metric.name: [] for metric in self.metrics}
        testing_metrics = {metric.name: [] for metric in self.metrics}
        self.setup()
        with tqdm(epochs) as bar:
            for epoch in range(epochs):
                # TRAINING
                bar.set_description('TRAINING')
                for X, T  in train:
                    metrics = self.step(X, T, training=True)
                    for name, value in metrics.items():
                        training_metrics[name].append(value)
                # WRITING METRICS
                with self.writer_train.as_default():
                    for metric in self.metrics:
                        tf.summary.scalar(metric.name, metric.result(), step=self.prev_epochs)
                self.reset_metrics()
                # TESTING
                bar.set_description('TESTING')
                for X, T  in test:
                    metrics = self.step(X, T, training=False)
                    for name, value in metrics.items():
                        testing_metrics[name].append(value)
                # WRITING METRICS
                with self.writer_val.as_default():
                    for metric in self.metrics:
                        tf.summary.scalar(metric.name, metric.result(), step=self.prev_epochs)
                self.reset_metrics()
                bar.update(1)
                self.prev_epochs += 1
            metrics = {metric.name: [training_metrics[metric.name], testing_metrics[metric.name]] for metric in self.metrics}
            return metrics
```

Ton ensure that the comparison is meaningful I will use the identical encoder and decoder architecture for the vanilla autoencoder as well as the variatioanl autoencoder.

```python
class EncoderBlock(tf.keras.layers.Layer):
    def __init__(self, filters_in, filters_out, **kwargs):
        super().__init__(**kwargs)
        self.conv_0 = tf.keras.layers.Conv2D(filters=filters_in, kernel_size=(3, 3), 
                                             strides=(1, 1), padding='same', activation='relu')
        self.norm_0 = tf.keras.layers.BatchNormalization()
        self.conv_1 = tf.keras.layers.Conv2D(filters=filters_in, kernel_size=(3, 3), 
                                             strides=(1, 1), padding='same', activation='relu')
        self.norm_1 = tf.keras.layers.BatchNormalization()
        self.conv_2 = tf.keras.layers.Conv2D(filters=filters_out, kernel_size=(3, 3), 
                                             strides=(2, 2), padding='same', activation='relu')
        self.norm_2 = tf.keras.layers.BatchNormalization()
    
    
    def call(self, x, training=False):
        z = self.conv_0(x)
        z = self.norm_0(z, training=training)
        z = self.conv_1(z)
        z = self.norm_1(z, training=training)
        z = z + x
        z = self.conv_2(z)
        z = self.norm_2(z, training=training)
        return z



class Encoder(Model):
    def __init__(self, units, **kwargs):
        super().__init__(**kwargs)
        # BLOCK ONE: shape=(28 -> 14), units=(100352 -> 50176)
        self.block_0 = EncoderBlock(128, 256)
        # BLOCK TWO: shape=(14 -> 7), units=(50176 -> 25088)
        self.block_1 = EncoderBlock(256, 512)
        # BLOCK THREE: shape=(7 -> 4), units=(25088 -> 8192)
        self.block_2 = EncoderBlock(512, 512)
        # BLOCK THREE: shape=(4 -> 2), units=(8192 -> 2048)
        self.block_3 = EncoderBlock(512, 512)
        # AVERAGING: units=(2048 -> 512)
        self.conv    = tf.keras.layers.Conv2D(filters=512, kernel_size=(2, 2), 
                                strides=(1, 1), padding='same', activation='relu')
        # COMPRESSING: units=(512 -> units)
        self.flatten = tf.keras.layers.Flatten()
        self.dense   = tf.keras.layers.Dense(units)


    def call(self, x, training=False):
        x = self.block_0(x, training=training)
        x = self.block_1(x, training=training)
        x = self.block_2(x, training=training)
        x = self.block_3(x, training=training)
        x = self.conv(x)
        x = self.flatten(x)
        x = self.dense(x)
        return x



class DecoderBlock(tf.keras.layers.Layer):
    def __init__(self, filters, **kwargs):
        super().__init__(**kwargs)
        self.conv_0 = tf.keras.layers.Conv2DTranspose(filters=filters, kernel_size=(3, 3), 
                                            strides=(2, 2), padding='same', activation='relu')
        self.norm_0 = tf.keras.layers.BatchNormalization()
        self.conv_1 = tf.keras.layers.Conv2DTranspose(filters=filters, kernel_size=(3, 3), 
                                            strides=(1, 1), padding='same', activation='relu')
        self.norm_1 = tf.keras.layers.BatchNormalization()
        self.conv_2 = tf.keras.layers.Conv2DTranspose(filters=filters, kernel_size=(3, 3), 
                                            strides=(1, 1), padding='same')#, activation='relu')
        self.norm_2 = tf.keras.layers.BatchNormalization()
    

    def call(self, x, training=False):
        x = self.conv_0(x)
        x = self.norm_0(x, training=training)
        z = self.conv_1(x)
        z = self.norm_1(z, training=training)
        z = self.conv_2(z)
        z = self.norm_2(z, training=training)
        z = z + x
        return z



class Decoder(Model):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        # DECOMPRESSING: units=(units -> 512)
        self.dense   = tf.keras.layers.Dense(512, activation='relu')
        # RESHAPING: (512 -> (2, 2, 256))
        self.reshape = tf.keras.layers.Reshape((1, 1, 512))
        # AVERAGING: units=(256 -> 1024)
        self.conv_0  = tf.keras.layers.Conv2DTranspose(filters=512, kernel_size=(2, 2), 
                                            strides=(2, 2), padding='same', activation='relu')
        # BLOCK ONE: shape=(2 -> 4), units=(1024 -> 8192) 
        self.block_0 = DecoderBlock(512)
        # BLOCK ONE: shape=(4 -> 7), units=(8192 -> 25088) 
        self.block_1 = DecoderBlock(512)
        self.crop    = tf.keras.layers.Cropping2D(cropping=((0, 1), (0, 1)))
        # BLOCK TWO: shape=(7 -> 14), units=(25088 -> 50176)
        self.block_2 = DecoderBlock(256)
        # BLOCK THREE: shape=(14 -> 28), units=(50176 -> 100352)
        self.block_3 = DecoderBlock(128)
        # RECONSTRUCTION: shape=(28 -> 28), units=(100352, 784)
        self.conv_1  = tf.keras.layers.Conv2DTranspose(filters=1, kernel_size=(3, 3), 
                                strides=(1, 1), padding='same', activation='sigmoid')


    def call(self, x, training=False):
        x = self.dense(x)
        x = self.reshape(x)
        x = self.conv_0(x)
        x = self.block_0(x, training=training)
        x = self.block_1(x, training=training)
        x = self.crop(x)
        x = self.block_2(x, training=training)
        x = self.block_3(x, training=training)
        x = self.conv_1(x)
        return x



class AutoEncoder(Model):
    NAME = 'Vanilla Autoencoder'
    def __init__(self, units, **kwargs):
        super().__init__(**kwargs)
        self.encoder = Encoder(units)
        self.decoder = Decoder()
        self.mse = tf.keras.losses.MeanSquaredError()
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)


    @tf.function
    def call(self, x, training=False):
        z = self.encoder(x, training=training)
        z = self.decoder(z, training=training)
        return z


    @tf.function
    def loss(self, x, y, training=False):
        x = self(x, training=True)
        loss = self.mse(x, y)
        return loss


    def display(self, data):
        x, y = next(iter(data))
        z = self(x)

        x = tf.concat([x[i, ...] for i in range(4)], axis=1)
        plt.imshow(x, cmap='binary')
        plt.title('Input')
        plt.show()
        y = tf.concat([y[i, ...] for i in range(4)], axis=1)
        plt.imshow(y, cmap='binary')
        plt.title('Target')
        plt.show()
        z = tf.concat([z[i, ...] for i in range(4)], axis=1)
        plt.imshow(z, cmap='binary')
        plt.title('Reconstruction')
        plt.show()



class VariationalAutoEncoder(AutoEncoder):
    NAME = 'Variational Autoencoder'
    def __init__(self, units, **kwargs):
        super().__init__(units, **kwargs)
        self.units = units
        self.encoder = Encoder(units * 2)
        self.decoder = Decoder()
        self.mse = tf.keras.losses.MeanSquaredError()
        self.bce = tf.keras.losses.BinaryCrossentropy()
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
        self.alpha = 1e-3


    @tf.function
    def call(self, x, training=False):
        z = self.encoder(x, training=training)
        z, mu, sigma = self.sample(z)
        x = self.decoder(z, training=training)
        if training:
            return x, mu, sigma
        else:
            return x


    @tf.function
    def sample(self, z):
        batch_size = z.shape[0]
        noise = tf.random.normal(shape=(batch_size, self.units))
        mu    = z[...,:self.units]
        sigma = z[...,self.units:]
        z = sigma * noise + mu
        return z, mu, sigma


    @tf.function
    def loss(self, x, y, training=False):
        x, mu, sigma = self(x, training=True)
        DKL = 1/2 * tf.reduce_mean(sigma**2 + mu**2 - 1 - tf.math.log(1e-12 + sigma**2))
        MSE = tf.reduce_mean((x - y)**2)
        loss = MSE + DKL * self.alpha
        return loss
```

# Autoencoder

```python
AE = AutoEncoder(32)
AE.display(test)
```

```python
AE.train(train, test, 10)
AE.display(test)
```

# Variational Autoencoder

```python
VAE = VariationalAutoEncoder(32)
VAE.display(test)
```

Too big of an alpha leads to no reconstruction at all and small alpha do not really have an effect. Therefore I adjust the alpha (DKL contribution to the loss) durin runtime.

```python
VAE.alpha = 1.0
VAE.train(train, test, 10)
VAE.display(test)
```
