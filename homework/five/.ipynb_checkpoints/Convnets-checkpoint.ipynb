```python
import tensorflow_datasets as tfds
import tensorflow as tf
import numpy as np
import math

(train, test) , info = tfds.load('mnist', split =['train', 'test'], 
                                 as_supervised=True , with_info=True)
```

```python
def preprocess(data):
    # DATA TYPE
    data = data.map(lambda img, target: (tf.cast(img, tf.float32), target))
    # NORMALIZE
    data = data.map(lambda img, target: (img/128. - 1., target))
    # ONE HOT ENCODING
    data = data.map(lambda img, target: (img, tf.one_hot(target, depth=10)))
    # DATAFLOW PREP
    data = data.cache()
    data = data.shuffle(1000, seed=42)
    data = data.batch(32)
    data = data.prefetch(20)
    return data

train = train.apply(preprocess)
test  = test.apply(preprocess)
```

```python
from tqdm.notebook import tqdm
import datetime


class DeepGuck(tf.keras.Model):
    L2 = tf.keras.regularizers.L2(1e-3)
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.conv_1 = tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3), 
                                             strides=(1, 1), activation='relu', 
                                             kernel_regularizer=self.L2,
                                             padding='same') # (28, 28) -> (28, 28)
        self.pool_1 = tf.keras.layers.MaxPooling2D(pool_size=(2, 2), padding="valid")# (28, 28) -> (14, 14)
        self.conv_2 = tf.keras.layers.SeparableConv2D(filters=64, kernel_size=(3, 3), 
                                                      strides=(2, 2), activation='relu', 
                                                      kernel_regularizer=self.L2,
                                                      depthwise_regularizer=self.L2,
                                                      padding='same') # (14, 14) -> (7, 7)
        self.conv_3 = tf.keras.layers.SeparableConv2D(filters=96, kernel_size=(3, 3), 
                                                      strides=(1, 1), activation='relu', 
                                                      kernel_regularizer=self.L2,
                                                      depthwise_regularizer=self.L2,
                                                      padding='valid') # (7, 7) -> (5, 5)
        self.conv_4 = tf.keras.layers.SeparableConv2D(filters=128, kernel_size=(3, 3), 
                                                      strides=(1, 1), activation='relu', 
                                                      kernel_regularizer=self.L2,
                                                      depthwise_regularizer=self.L2,
                                                      padding='valid') # (5, 5) -> (3, 3)
        self.conv_5 = tf.keras.layers.Conv2D(filters=128, kernel_size=(3, 3), 
                                             strides=(1, 1), activation='relu', 
                                             kernel_regularizer=self.L2,
                                             padding='valid') # (3, 3) -> (1, 1)
        #self.pool_2= tf.keras.layers.AveragePooling2D(pool_size=(3, 3), padding="valid") # (3, 3) -> (1, 1)
        self.flatten = tf.keras.layers.Flatten()
        self.affine_1 = tf.keras.layers.Dense(64, activation='relu', kernel_regularizer=self.L2)
        self.affine_2 = tf.keras.layers.Dense(48,  activation='relu', kernel_regularizer=self.L2)
        self.affine_3 = tf.keras.layers.Dense(10,  activation='softmax', kernel_regularizer=self.L2)
        self.loss = tf.keras.losses.CategoricalCrossentropy()
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
        self.accuracy_metric = tf.keras.metrics.CategoricalAccuracy(name="acc")
        self.loss_metric = tf.keras.metrics.Mean(name="loss")
    
    
    def reset_metrics(self):
        for metric in self.metrics:
            metric.reset_state()
    
    
    def setup(self):
        # DEFINE PATHS
        current_time = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
        train_log_path = f"logs/{current_time}/train"
        val_log_path = f"logs/{current_time}/val"
        # CONSTRUCT WRITERS
        self.writer_train = tf.summary.create_file_writer(train_log_path)
        self.writer_val   = tf.summary.create_file_writer(val_log_path)
    
    
    @tf.function
    def call(self, x):
        for layer in self.layers:
            x = layer(x)
        return x
    
    
    @tf.function
    def step(self, X, T, train=True):
        if train:
            with tf.GradientTape() as tape:
                Y = self(X)
                L = self.loss(T, Y)
            gradient = tape.gradient(L, self.trainable_weights)
            self.optimizer.apply_gradients(zip(gradient, self.trainable_weights))
        else:
            Y = self(X)
            L = self.loss(T, Y)
        self.loss_metric.update_state(L)
        self.accuracy_metric.update_state(T, Y)
        return {metric.name: float(metric.result()) for metric in self.metrics}
    
    
    def train(self, train, test, epochs):
        training_metrics = {metric.name: [] for metric in self.metrics}
        testing_metrics = {metric.name: [] for metric in self.metrics}
        self.setup()
        with tqdm(epochs) as bar:
            for epoch in range(epochs):
                # TRAINING
                bar.set_description('TRAINING')
                for X, T  in train:
                    metrics = self.step(X, T)
                    for name, value in metrics.items():
                        training_metrics[name].append(value)
                with self.writer_train.as_default():
                    for metric in model.metrics:
                        tf.summary.scalar(metric.name, metric.result(), step=epoch)
                self.reset_metrics()
                # TESTING
                bar.set_description('TESTING')
                for X, T  in test:
                    metrics = self.step(X, T, train=False)
                    for name, value in metrics.items():
                        testing_metrics[name].append(value)
                with self.writer_train.as_default():
                    for metric in model.metrics:
                        tf.summary.scalar(metric.name, metric.result(), step=epoch)
                self.reset_metrics()
                bar.update(1)
            metrics = {metric.name: [training_metrics[metric.name], testing_metrics[metric.name]] for metric in self.metrics}
            return metrics



model = DeepGuck()
print(model(np.ones((1, 28, 28, 1))))
model.summary()
```

```python
def smooth(data, level):
    assert level > 1
    data = np.array([data[0]] * (level//2) + data + [data[-1]] * (level//2 - ((level + 1)%2)))
    kernel = np.array([1/level] * level)
    return np.convolve(kernel, data, 'valid')
```

An accurcy of $98.9%$ makes the previous model sufficent. So rather then improving the accuracy I want to motivate the second models design by curriosity. Specifically I would like to see how many conv layers we can stack ontop of each other until the whole thing becomes untrainable. 
https://github.com/Spinkk/BatchEntropyRegularization introduce batch entropy maximization to ensure, that information about the underlying distribution is propagated for through each layer. I basically did the same but added filter and feature map entropy.

The activation patterns in each unit seem to follow a normal distribution. From this we can derive the entropy of that distribution to be:
$$H(X) = \frac{1}{2}\log(\tau e \sigma(X)^2)$$ where $\tau$ is the circle constant, $e$ the natural lgarthms base and $\sigma$ the standart deviation observed along the batch/filter or feature map.

```python
X, _ = batch = next(iter(test))
N = 5

for layer in model.layers:
    if isinstance(layer, tf.keras.layers.MaxPooling2D) or isinstance(layer, tf.keras.layers.Flatten):
        X = layer(X)
        continue
    activation = layer.activation
    layer.activation = tf.keras.activations.get(None)
    X = layer(X)
    layer.activation = activation
    if isinstance(layer, tf.keras.layers.Dense):
        continue
    fig, ax = plt.subplots(1, N, sharey=True, tight_layout=True)
    for i in range(N):
        ax[i].hist(tf.reshape(X[:,:,:,i], (-1,)))
    plt.show()
```

```python
sep_conv = tf.keras.layers.SeparableConv2D
conv     = tf.keras.layers.Conv2D



class EntropyConv(tf.keras.layers.Layer):
    def __init__(self, layer_class, filters, kernel_size=(3, 3), 
                 strides=(2, 2), activation='relu', padding='valid', **kwargs):
        super().__init__(**kwargs)
        self.layer = layer_class(filters=filters, kernel_size=kernel_size, 
                                 strides=strides, padding=padding, 
                                 kernel_regularizer=tf.keras.regularizers.L2(1e-1),
                                 **kwargs)
        self.activation = tf.keras.activations.get(activation)
    
    
    @tf.function
    def Entropy(self, z, axis):
        var = tf.math.reduce_std(z, axis=axis)**2
        I = 1/2 * tf.math.log(np.math.tau * np.math.e * var + 1)
        H = tf.math.reduce_mean(I)
        return H
    
    
    @tf.function
    def call(self, x, training=False):
        z = self.layer(x)
        if training:
            H_batch  = self.Entropy(z, axis=0)
            H_filter = self.Entropy(z, axis=3)
            H_image  = self.Entropy(z, axis=(1, 2))
        a = self.activation(z)
        if training:
            return a, H_batch, H_filter, H_image
        else:
            return a



class EntropyMaximizer(tf.keras.Model):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        for i in range(100):
            layer = EntropyConv(conv, 32, strides=(1, 1), padding='same')
            self.__setattr__(f'conv_{i}', layer)
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
        self.loss_metric = tf.keras.metrics.Mean(name="loss")
        
        
    def reset_metrics(self):
        for metric in self.metrics:
            metric.reset_state()
    
    
    def setup(self):
        # DEFINE PATHS
        current_time   = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
        train_log_path = f"logs/entropy_{current_time}/train"
        val_log_path   = f"logs/entropy_{current_time}/val"
        # CONSTRUCT WRITERS
        self.writer_train = tf.summary.create_file_writer(train_log_path)
        self.writer_val   = tf.summary.create_file_writer(val_log_path)
    
    
    @tf.function
    def call(self, x):
        for layer in self.layers:
            x = layer(x)
        return x
    
    
    @tf.function
    def loss(self, x):
        H_batch  = 0.
        H_filter = 0.
        H_image  = 0.
        for layer in self.layers:
            x, h_batch, h_filter, h_image = layer(x, training=True)
            H_batch  += 1 * (h_batch/4  - 1)**2
            H_filter += 1 * (h_filter - 1)**2
            H_image  += 1 * (h_image/1 - 1)**2
        return H_batch + 0*H_filter + H_image



entropy_base = EntropyMaximizer()
print(tf.reshape(entropy_base(np.ones((1, 28, 28, 1))), (-1,)))
entropy_base.summary()
```

```python
class EntropyClassifier(tf.keras.Model):
    def __init__(self, base, **kwargs):
        super().__init__(**kwargs)
        self.base = base
        #self.base.trainable = False
        self.flatten = tf.keras.layers.Flatten()
        self.affine = tf.keras.layers.Dense(10, activation='softmax')
        self.optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)
        self.loss_metric = tf.keras.metrics.Mean(name="loss")
        self.ce = tf.keras.losses.CategoricalCrossentropy()
    
    
    def loss(self, T, X):
        Y = self(X)
        CE = self.ce(T, Y)
        BE = self.base.loss(X)
        return CE + BE * CE
    
    
    def reset_metrics(self):
        for metric in self.metrics:
            metric.reset_state()
    
    
    def setup(self):
        # DEFINE PATHS
        current_time   = datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
        train_log_path = f"logs/entropy_{current_time}/train"
        val_log_path   = f"logs/entropy_{current_time}/val"
        # CONSTRUCT WRITERS
        self.writer_train = tf.summary.create_file_writer(train_log_path)
        self.writer_val   = tf.summary.create_file_writer(val_log_path)
    
    
    @tf.function
    def call(self, x):
        x = self.base(x)
        x = self.flatten(x)
        x = self.affine(x)
        return x
    
    
    @tf.function
    def step(self, X, T, train=True):
        if train:
            with tf.GradientTape() as tape:
                L = self.loss(T, X)
            gradient = tape.gradient(L, self.trainable_weights)
            self.optimizer.apply_gradients(zip(gradient, self.trainable_weights))
        else:
            L = self.loss(T, X)
        self.loss_metric.update_state(L)
        #self.accuracy_metric.update_state(T, Y)
        return {metric.name: float(metric.result()) for metric in self.metrics}
    
    
    def train(self, train, test, epochs):
        training_metrics = {metric.name: [] for metric in self.metrics}
        testing_metrics  = {metric.name: [] for metric in self.metrics}
        self.setup()
        with tqdm(epochs) as bar:
            for epoch in range(epochs):
                # TRAINING
                bar.set_description('TRAINING')
                for X, T  in train:
                    metrics = self.step(X, T)
                    for name, value in metrics.items():
                        training_metrics[name].append(value)
                with self.writer_train.as_default():
                    for metric in model.metrics:
                        tf.summary.scalar(metric.name, metric.result(), step=epoch)
                self.reset_metrics()
                # TESTING
                bar.set_description('TESTING')
                for X, T  in test:
                    metrics = self.step(X, T, train=False)
                    for name, value in metrics.items():
                        testing_metrics[name].append(value)
                with self.writer_train.as_default():
                    for metric in model.metrics:
                        tf.summary.scalar(metric.name, metric.result(), step=epoch)
                self.reset_metrics()
                bar.update(1)
            metrics = {metric.name: [training_metrics[metric.name], testing_metrics[metric.name]] for metric in self.metrics}
            return metrics



entropy_classifier = EntropyClassifier(entropy_base)
print(tf.reshape(entropy_classifier(np.zeros((1, 28, 28, 1))), (-1,)))
entropy_classifier.summary()
```
